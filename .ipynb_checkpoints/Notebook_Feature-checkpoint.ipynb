```python
import numpy as np
import load_data
import clean_data
import transform_data
import feature_selection
```

```python
#Aplicando las funciones de la tarea anterior
datos=load_data.carga_archivo('consumo-agua.csv')
load_data.observaciones_variables(datos)
clean_data.estandariza_variables(datos)
transform_data.tipo_variables(datos)
transform_data.separar_variable(datos,'geo_point','latitud','longitud',',')
transform_data.eliminar_variable(datos,'geo_point')
transform_data.eliminar_variable(datos,'geo_shape')
transform_data.cambiar_tipo_variable(datos,'latitud','float64')
transform_data.cambiar_tipo_variable(datos,'longitud','float64')
transform_data.cambiar_tipo_variable(datos,'bimestre','str')
transform_data.cambiar_tipo_variable(datos,'gid','str')
transform_data.cambiar_minusculas_variable(datos,'colonia')
transform_data.cambiar_minusculas_variable(datos,'alcaldia')
transform_data.cambiar_minusculas_variable(datos,'indice_des')
transform_data.quitar_acentos(datos)
```

```python
transform_data.eliminar_variable(datos,'nomgeo') #dado que se ocupara alcaldía
transform_data.eliminar_variable(datos,'gid') #dado que es una llave
transform_data.eliminar_variable(datos,'anio') #dado que todo es 2019 y no tiene variacion
transform_data.eliminar_variable(datos,'colonia') #dado que en su lugar se ocupará alcaldía
columna_mover = ['consumo_total']
nuevo_orden = np.hstack((datos.columns.difference(columna_mover), columna_mover))
datos_nuevo_orden = datos.reindex(columns=nuevo_orden)
datos_nuevo_orden.head()
```

```python
#Prueba de la función chi_squared
feature_selection.chi_squared(datos_nuevo_orden,'consumo_total',0.05)
```

```python
x = datos_nuevo_orden.iloc[:, :-1]
y = datos_nuevo_orden.iloc[:, -1].values
x.head()
```

```python
#Prueba de la función backward_selection
feature_selection.backward_selection(x,y,0.05)
```
